# 3. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# 12. dtypes parameter
### dtype → specifies the data type for columns while reading the file (e.g., int, float, str)

# 13. Handling Dates
### parse_dates → converts date columns (usually read as strings/objects) into datetime format so you can access day, month, year, etc.
#### obsever the date in the dataset

```python
pd.read_csv('IPL Matches 2008-2020.csv').info()
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

# 14. Convertors
### converters → applies custom functions to transform specific column values while reading the file

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

# 15. na_values parameter

# 16. Loading a huge dataset in chunks

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
