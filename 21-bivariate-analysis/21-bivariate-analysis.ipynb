# 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue='sex', style='smoker', size='size')
```

# 2. Bar Plot (Numerical - Categorical)

```python
titanic.sample(2)
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Sex')
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

# 4. Distplot (Numerical - Categorical)

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived', fill=False)
```

# 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100 #if want to plot use:plot(kind"")
```

# 6. ClusterMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# 8. Lineplot (Numerical - Numerical)
### Use a lineplot when you want to show how a numeric value changes over a continuous or ordered variable (usually time).

```python
flights.sample(6)
```

```python
new = flights.groupby('year', as_index=False).sum(numeric_only=True)
```

### Groups data by year and calculates the total (sum) of numerical columns for each year.

```python
new
```

```python
sns.lineplot(x='year', y='passengers', data=new)
```

```python
flights.pivot_table(values='passengers', index='month', columns='year') 
# Creates a table with months as rows, years as columns, and passenger counts as values
```

```python
sns.heatmap(
    flights.pivot_table(values='passengers', index='month', columns='year', observed=False)
)
```

```python
sns.clustermap(    flights.pivot_table(values='passengers', index='month', columns='year', observed=False))
```
