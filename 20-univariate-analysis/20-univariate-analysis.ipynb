## Boxplot

## 📦 Box Plot Summary

**Five-number summary:**
- Min
- Q1 (25%)
- Median (Q2, 50%)
- Q3 (75%)
- Max

**Formulas:**
- IQR = Q3 − Q1
- Lower Bound = Q1 − 1.5 × IQR
- Upper Bound = Q3 + 1.5 × IQR

**Outliers:**
- < Lower Bound or > Upper Bound

**Components:**
- Box → Q1 to Q3
- Line → Median
- Whiskers → Min & Max (no outliers)
- Points → Outliers

**Interpretation:**
- Bigger box → more spread
- Median center → symmetric
- Longer whisker → skew

```python
sns.boxplot(df['Age'])
```
